```python
from ks.KS_environment import KSenv
import numpy as np

burn_in = 1500
env = KSenv(nu=0.08,
            actuator_locs=np.linspace(0.2, 2 * np.pi - 0.2, 7),#np.array([0.0, np.pi / 4, np.pi / 2, 3 * np.pi / 4]),
            # sensor_locs=np.array([0, 2 * np.pi, 64]),
            burn_in=burn_in)
```

```python
_, u_hist = env.burnin_reset()
```

```python
u_hist.shape
```

```python
import matplotlib.pyplot as plt

x = np.arange(64)
y = np.arange(burn_in)

X,Y = np.meshgrid(x,y)

plt.figure()
plt.contourf(X,Y,u_hist)
plt.xlabel("x space")
plt.ylabel("Timesteps")
```

```python
plt.figure()
plt.plot(np.arange(u_hist.shape[0]),0.5*np.linalg.norm(u_hist, axis = -1))
plt.xlabel("Timesteps")
plt.ylabel("Energy")
```
